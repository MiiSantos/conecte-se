```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('pontos_wifi.csv',  encoding='latin-1', sep=';', skiprows=0, low_memory=False)
```

```python
df
```

```python
new_header = df.iloc[0]  # Pega a primeira linha
df = df[1:]  # Remove a primeira linha do DataFrame
df.columns = new_header  # Define os novos cabeçalhos
```

```python
df.columns
```

```python
contagem_pontos = df.groupby('Bairro').size().reset_index(name='Pontos de Acesso')
```

```python
contagem_pontos
```

```python
contagem_pontos_ordenada = contagem_pontos.sort_values(by='Pontos de Acesso', ascending=False)

```

```python
contagem_pontos_ordenada
```

```python
top_5_bairros = contagem_pontos_ordenada.head(10)

# Plotar o gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(top_5_bairros['Pontos de Acesso'], labels=top_5_bairros['Bairro'], autopct='%1.1f%%', colors=plt.get_cmap('Set3').colors)
plt.title('Top 5 Bairros com Maior Número de Pontos de Acesso')
plt.show()
```
